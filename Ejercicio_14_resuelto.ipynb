```python
#Ejercicio 14 - Samuel LÃ³pez
import math

class Calculos:
  @staticmethod
  def calcular_cuadr(numero):
    return math.pow(numero, 2)

  @staticmethod
  def calcular_cubo(numero):
    return math.pow(numero, 3)

numero = 4.0
cuadrado = Calculos.calcular_cuadr(numero)
cubo = Calculos.calcular_cubo(numero)

print("Cuadrado:" , cuadrado)
print("Cubo" , cubo)
```
